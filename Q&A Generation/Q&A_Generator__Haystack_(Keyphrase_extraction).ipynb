##**TEXT-PREPROCESSING**

```python
def remove_pg_num(text):
    text = re.sub('[[]\d+[]]|[[]\d[]]', '', text)
    # text = re.sub('[[][0-9][0-9][0-9][]]', '', text)
    return text
```

```python
def remove_extra_spaces(text):
  text = re.sub('\s+', ' ', text)
  return text
```

```python
def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
```

```python
def remove_ar(text):
    text = re.sub('[أ-ي]', '', text)
    return text
```

```python
def remove_num(text):
    text = re.sub('[0-9]', '', text)
    return text
```

```python
def remove_mails(text):
    text = re.sub(r'[A-Za-z0-9]*@[A-Za-z]*\.?[A-Za-z0-9]*', "", text)
    return text
```

```python
def remove_dot(text):
    text = re.sub(r'\.+', ".", text)
    text = re.sub(r'\,+', ",", text)
    return text
```

```python
def remove_punctuations(text):
    punctuations = '!"#$%&\()*/:;<=>?@[\\]^_`{|}~-'
    return text.translate(str.maketrans('', '', punctuations))
```

```python
def remove_single_quotes(text):
    text = re.sub("'", '', text)
    return text
```

**TEXT-PREPROCESSING (FUNCTION)**

```python
# The parameter list is list of text-lines
def text_preprocessing(list):
  #if there is any (\n) in the end of each line
  for line in list:
    line = line.strip()
  #To concatenate all text-data
  data = " ".join([line for line in list])

  data = remove_pg_num(data)
  data = remove_url(data)
  data = remove_ar(data)
  data = remove_mails(data)
  data = contractions.fix(data)
  data = remove_punctuations(data)
  data = remove_num(data)
  data = remove_dot(data)
  data = remove_single_quotes(data)
  data = remove_extra_spaces(data)

  return data

  
  
```

```python
text_preprocessed2 = text_preprocessing(data)

text_preprocessed2
```

##**Keyphrase Sentence Map**

```python
def keyphrase_sentence_map(text, keyphrases):

  regex = re.compile('\,|\.')
  sentences = regex.split(text)
  
  #  Find all sentences
  sentences_map = {}
  for keyphrase in keyphrases:
    for sentence in sentences:
      if sentence.find(keyphrase)!= -1:
        if keyphrase in sentences_map.keys():
          sentences_map[keyphrase].append(sentence)
        else:
          sentences_map[keyphrase] = [sentence]

  #  Concatenate all sentences
  for keyphrase in keyphrases:
    sentences_map[keyphrase]= ", ".join([sentence for sentence in sentences_map[keyphrase] ])

  return sentences_map
```

```python
docs =[]
for sentence in sentences_map.values():
  docs.append({"content": sentence})

# Initialize document store and write in the documents
document_store = ElasticsearchDocumentStore()
document_store.delete_documents()
document_store.write_documents(docs)

# Initialize Question Generator
question_generator = QuestionGenerator()

```

```python
retriever = BM25Retriever(document_store=document_store)
rqg_pipeline = RetrieverQuestionGenerationPipeline(retriever, question_generator)
topic = 'Machine Learning'
print(f"\n * Generating questions for documents matching the query '%s'\n" % (topic))
q_result = rqg_pipeline.run(query=topic)
print_questions(q_result)
```

```python

a_results=[]
reader = FARMReader("deepset/roberta-base-squad2")
qag_pipeline = QuestionAnswerGenerationPipeline(question_generator, reader)
for idx, document in enumerate(tqdm(document_store)):

    print(f"\n * Generating questions and answers for document {idx}: {document.content[:100]}...\n")
    a_result = qag_pipeline.run(documents=[document])
    a_results.append(a_result)
    print_questions(a_result)

```

```python
c_q_a = {}
c =[]
q =[]
a =[]

for list_ in a_results:
  for q_ , a_, c_ in zip(list_['queries'], list_['answers'], list_['answers']):
    q.append(q_)
    a.append(a_[0].answer)
    c.append(c_[0].context)

c_q_a["Context"]= c
c_q_a["Question"]= q
c_q_a["Haystack_Answer"]= a
```
