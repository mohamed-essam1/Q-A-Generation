##**Load Data**

```python
import pandas as pd

df = pd.read_csv('Q&A Generator.csv')
df
```

##**Question-generation & Question-answering(By patil-suraj Repo)T5 model**

```python
from transformers import pipeline
question_answerer = pipeline("question-answering", model='distilbert-base-uncased-distilled-squad')

```

```python
DistilBert_Answers = []
for q, c in zip(df['Question'], df['Context']):
    DistilBert_Answers.append(question_answerer(question=q, context=c)['answer'])

df['DistilBert_Answer'] = DistilBert_Answers
```

##**Save Data in The CSV File**

```python
# Save dataframe as csv file in the current folder
df.to_csv('/content/Q&A Generator.csv', index = False, encoding='utf-8') # False: not include index
```
