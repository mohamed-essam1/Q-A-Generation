```python
from pipelines import pipeline
nlp_ans = pipeline("multitask-qa-qg", model="valhalla/t5-base-qa-qg-hl")
```

```python
T5_Answers = []
for q, c in zip(df['Question'], df['Context']):
    T5_Answers.append(nlp_ans({"question": q,"context": c}))

df['T5_Answer'] = T5_Answers
```
