```python
from allennlp.predictors.predictor import Predictor
import allennlp_models.rc

predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/bidaf-elmo.2021-02-11.tar.gz")
```

```python
allen_ans = []

for q, c in zip(df['Question'], df['Context']):
   allen_ans.append(predictor.predict(
      passage=c,
      question= q
  ))
   
df['AllenNLP_Answer'] = allen_ans
```

##**Save Data in The CSV File**

```python
# Save dataframe as csv file in the current folder
df.to_csv('Q&A Generator.csv', index = False, encoding='utf-8') # False: not include index
```
